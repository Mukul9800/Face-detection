```python
class JHUCrowdDataset(Dataset):
    def __init__(self, root, transforms=None):
        self.root = root
        self.transforms = transforms
        self.images = list(sorted(os.listdir(os.path.join(root, "images"))))
        self.labels = list(sorted(os.listdir(os.path.join(root, "labels"))))

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = os.path.join(self.root, "images", self.images[idx])
        label_path = os.path.join(self.root, "labels", self.labels[idx])
        img = Image.open(img_path).convert("RGB")
        with open(label_path) as f:
            boxes = []
            for line in f:
                label = line.strip().split()
                x_center, y_center, width, height = map(float, label[1:])
                x_min = (x_center - width / 2) * img.width
                y_min = (y_center - height / 2) * img.height
                x_max = (x_center + width / 2) * img.width
                y_max = (y_center + height / 2) * img.height
                boxes.append([x_min, y_min, x_max, y_max])
            num_objs = len(boxes)
            boxes = torch.as_tensor(boxes, dtype=torch.float32)
            labels = torch.ones((num_objs,), dtype=torch.int64)  # all labels are '1' (face)
            target = {}
            target["boxes"] = boxes
            target["labels"] = labels

        if self.transforms is not None:
            img = self.transforms(img)

        return img, target

def get_transform(train):
    transforms_list = [transforms.ToTensor()]
    return transforms.Compose(transforms_list)
```

```python
train_dataset = JHUCrowdDataset('/content/rishav/JHU_Crowd/YOLO_format/train', get_transform(train=True))
valid_dataset = JHUCrowdDataset('/content/rishav/JHU_Crowd/YOLO_format/valid', get_transform(train=False))
test_dataset = JHUCrowdDataset('/content/rishav/JHU_Crowd/YOLO_format/test', get_transform(train=False))

train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, num_workers=4, collate_fn=lambda x: tuple(zip(*x)))
valid_loader = DataLoader(valid_dataset, batch_size=1, shuffle=False, num_workers=4, collate_fn=lambda x: tuple(zip(*x)))
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False, num_workers=4, collate_fn=lambda x: tuple(zip(*x)))
```

```python
# Load a pre-trained model for classification and return only the features
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)

# Replace the classifier with a new one, that has num_classes which is user-defined
num_classes = 2  # 1 class (face) + background

# get number of input features for the classifier
in_features = model.roi_heads.box_predictor.cls_score.in_features

# replace the pre-trained head with a new one
model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

# Move model to the appropriate device
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)
```

```python
# Visualization function
def visualize(image, boxes):
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    ax = plt.gca()
    for box in boxes:
        x_min, y_min, x_max, y_max = box
        rect = plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, fill=False, color='red')
        ax.add_patch(rect)
    plt.show()

model.eval()
with torch.no_grad():
    for images, targets in test_loader:
        images = list(img.to(device) for img in images)
        outputs = model(images)
        img = np.array(images[0].cpu().permute(1, 2, 0))
        boxes = outputs[0]['boxes'].cpu().numpy()
        visualize(img, boxes)
```
